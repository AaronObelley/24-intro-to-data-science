```python
import pandas

df = pandas.read_csv('GDP_Sector_Income_2019_2023_30Cities.csv')
df.head(1)
```

```python
df.describe()
```

```python
null_data = df.isnull().sum()
null_data
```

```python
city_by_gdp = df.groupby(['City'])['GDP (in billion $)'].sum().sort_values(ascending=False).reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(40, 20))
plt.bar(city_by_gdp['City'], city_by_gdp['GDP (in billion $)'], color='red')
plt.title('Indian Cities by GDP', fontsize=20)
plt.xlabel('City')
plt.xticks(fontsize = 8)
plt.ylabel('GDP (in billion $)')
plt.yticks(fontsize = 8)
```
